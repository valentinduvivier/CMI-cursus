#### Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ------
# import
# ------

inf_range = 4
sup_range = 10

data = np.zeros((10**3, 20, sup_range-inf_range))

for i in range(inf_range, sup_range):
    
    if i<9:
        strg = '0'+str(i+1)
    else:
        strg = str(i+1)
        
    filename = 'SU2_copy/bin/DESIGNS/DSN_0'+strg+'/DIRECT/history_direct.csv'
    stock = np.loadtxt(filename, delimiter=', ', skiprows=1, dtype=float)

    # ---
    
    size_raw, size_col = np.size(stock,0), np.size(stock,1)
    data[:size_raw, :size_col, i-inf_range] = stock
    
data=data # exp data that made the curve
```

Superposition Cd graphs for 10 first DSG

```python
inf_lim = 80 # limit nb of iterations we display

# --- 

iterations = data[:inf_lim,2,:]
CD = data[:inf_lim,7,:]

plt.figure(figsize=(8,4))

plt.plot(iterations, CD, '-', lw=3)

plt.title('CD as a function of iterations', size=15)
plt.xlabel('iterations', size=15)
plt.ylabel('CD', size=15)

plt.legend(['DSG'+str(i+1) for i in range(inf_range, sup_range)])
plt.grid(True)

# plt.savefig('5_first_DSG.pdf')
```

Superposition Cd graphs for 10 last DSG

```python
# ------
# import
# ------

inf_range = 0
sup_range = 1

data = np.zeros((10**3, 20, sup_range-inf_range))

for i in range(inf_range, sup_range):
    
    if i<9:
        strg = '0'+str(i+1)
    else:
        strg = str(i+1)
        
    filename = 'SU2_copy/bin/DESIGNS/DSN_0'+strg+'/DIRECT/history_direct.csv'
    stock = np.loadtxt(filename, delimiter=', ', skiprows=1, dtype=float)

    # ---
    
    size_raw, size_col = np.size(stock,0), np.size(stock,1)
    data[:size_raw, :size_col, i-inf_range] = stock
    
data=data # exp data that made the curve
```

```python
inf_lim = -80 # limit nb of iterations we display

# --- 

iterations = data[:inf_lim,2,:]
CD = data[:inf_lim,7,:]

# --- 

plt.figure(figsize=(14,7))

plt.plot(iterations, CD, '-', lw=5)

plt.title('CD as a function of iterations', size=25)
plt.xlabel('iterations', size=25)
plt.ylabel('CD', size=25)

plt.xticks(size=20)
plt.yticks(size=20)

plt.legend(['DSG'+str(i+1) for i in range(inf_range, sup_range)], prop={'size':20})
plt.grid(True)

plt.savefig('15_ref.pdf')
```
