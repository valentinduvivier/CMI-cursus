Libraries

```python
import numpy as np
import matplotlib.pyplot as plt

# ------
# import
# ------

filename = 'Q4/results_data.csv'
data = np.loadtxt(filename, delimiter=',', skiprows=1, dtype=float)

data=data # exp data that made the curve
```

Code

```python
#data
Gc = 3.E-3

Uimp = data[:, 0]
E_f  = data[:,-1]*Gc

# --- 

plt.figure(figsize=(10,5))

plt.plot(Uimp, E_f, '.')

plt.title(r'$E_f$ as a function of $U_{imp}$', size=15)
plt.xlabel(r'$U_{imp}$', size=15)
plt.ylabel(r'$E_f$', size=15)

plt.grid(True)

# plt.savefig('E_f_Uimp_refined.png')
```
