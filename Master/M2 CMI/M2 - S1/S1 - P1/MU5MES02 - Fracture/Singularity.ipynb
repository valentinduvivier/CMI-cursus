```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
eps = 1.E-5
lamd_i = np.linspace(eps, 1. -eps, 100)

ratio_mu_vector = np.array([.5*_ for _ in range(1,4)])
omega_vector    = np.array([np.pi/4]) #np.array([1./_ for _ in range(1,5)]) * np.pi

for ratio_mu in ratio_mu_vector:
    for omega in omega_vector:
        
        plt.figure()
        
        plt.plot(lamd_i, ratio_mu * np.tan(lamd_i*omega)          , '.', color='orange'   , label=r'w={:.2f}pi & ratio_mu={:.1f}'.format(omega/np.pi, ratio_mu))
        plt.plot(lamd_i,          - np.tan(lamd_i*(np.pi - omega)), '.', color='royalblue', label=r'w={:.2f}pi & ratio_mu={:.1f}'.format(omega/np.pi, ratio_mu))
        
        plt.axis((-.05, 1+.05, -10, +10))
        
        plt.grid(True)
        plt.legend()
```
