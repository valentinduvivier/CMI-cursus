```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
As  = 1.
T   = 130000 * 10**6.

M   = 5.98   * 10**34

rho = 1300.
G   = 6.67   * 10**-11
g   = 9.81

R   = 6371   * 10**3.
Rg  = 42164  * 10**3.
r   = 1      * 10**6

# ------------------

# Initialisation
r = np.linspace(.1, 10*Rg, 10**4)
A = np.zeros((np.size(r)))

A = As * np.exp(rho*g*R**2/T * (1/R - 1/r + (R**2 - r**2)/(2*Rg**3)))

# ------------------

plt.figure(figsize=(8,6))

plt.plot(r/Rg, A/As, '-')
plt.plot(np.linspace(1, 1, 10**2), np.linspace(-0., max(A/As)+.1, 10**2), 'r--')

plt.title('Cable\'s thickness over altitude - carbon nanotube')
plt.xlabel(r'$\dfrac{r}{R_g}$')
plt.ylabel('Taper ratio')

plt.grid('True')

plt.tight_layout()

# ------------------

plt.savefig('Thickness.png')
```
