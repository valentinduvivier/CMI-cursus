```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
### N = 2

## Data

# Beam
E = 200 # Young modulus[Pa]
S = 1   # surface [m²]
L = 10   # length [m]

# Mesh
N = 2 # nb point
x = np.linspace(0, L, N)

# Forces
p0 = 10 # [Pa]
F0 = 10 # [N]

# -----------------------------

## Exact solution
u = 1/(E*S) * (p0 * (L/np.pi)**2 * np.sin(np.pi*x/L) + (p0 * L/np.pi + F0)*x)

## Approached solution
K = E*S/L * np.array([[1, 1], [1, 4/3]])
F = p0*L/np.pi * np.array([[1],[1 - 4/np.pi**2]]) + F0*np.array([[1],[1]])
a = np.dot(np.linalg.inv(K), F)

phi   = np.array([[x/L], [(x/L)**2]])

u_app = np.dot(np.transpose(phi[:,0,:]), a)

# -----------------------------

fig, ax1 = plt.subplots()

ax1.plot(x, u, 'r-', label='exacte', lw=2, color="red")
ax1.plot(x, u_app, 'g--', label='approx', lw=2, color="green")

ax1.set_title(f'ana - approx at order {N}')
ax1.set_xlabel('x')
ax1.set_ylabel(r'u(x)')

plt.legend()

# -----------------------------

## Error calculation
L_inf = np.zeros((N))
for i in range(N):
    L_inf[i] = np.max(np.abs(u[i] - u_app[i]))
    
L_2   = np.zeros((N))
for i in range(N):
    L_2[i]   = np.sqrt(np.sum(np.abs(u[i] - u_app[i])**2))
    
# -----------------------------

ax2 = ax1.twinx()

ax2.plot(x, L_inf, '*', label='L_inf', lw=2, color="orange")
ax2.plot(x, L_2, '--', label='L_2', lw=2, color="yellow")

ax2.set_title(f'ana - approx at order {N}')
ax2.set_xlabel('x')
ax2.set_ylabel(r'e(x)')

plt.legend()
plt.grid(True)
```
