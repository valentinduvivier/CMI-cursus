```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# f(theta)

def f(theta, U):
    alpha = np.pi/8
    return U*np.sin(theta)*(1-theta/alpha)
```

```python
# psi

def psi(theta, x, y, U):
        
    r = np.sqrt(np.square(x) + np.square(y))

    return r*f(theta,U)
```

```python
U = 1.

theta = np.linspace(0, 2*np.pi, 200)

xx = np.linspace(0, 2, 200)
yy = np.linspace(0, 2, 200)

x, y = np.meshgrid(xx,yy)

#plt.plot(x, psi(theta, r), 'r')
```

```python
import math as m

fig = plt.figure(figsize=(8,5))
plt.contour(x, y, psi(theta, x, y, U), 15, colors='black')
```

```python
fig = plt.figure(figsize=(10,5))

ax  = fig.add_subplot(121)

plt.plot(theta, f(theta, U), 15, label=r'f($\theta$)')
plt.plot(np.array([np.pi/2 for i in range(len(theta))]), f(theta, U), label = r'$\theta = \pi/2$')

plt.title(r'y=f($\theta$)')
plt.xlabel(r'$\theta$')
plt.ylabel(r'f($\theta$)')

plt.legend()
plt.grid('True')
```
